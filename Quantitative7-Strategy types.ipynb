# 第七章 量化投资策略种类

量化投资的策略种类并不多，主流的就四种：股票量化、CTA、高频、套利。

## 7.1 股票量化

股票量化就是指多因子$\alpha$策略。那么疑问就来了，股票量化就没有其它策略了？明明是有的呀？比如配对交易？

### 7.1.1 配对交易

配对交易策略的基本原理是基于两个相关性较高的股票或者其他证券，如果在未来时期保持着良好的相关性，一旦两者之间出现了背离的走势，且这种背离在未来是会得到纠正的，那么就可能产生套利的机会。对于配对交易的实践而言，如果两个相关性较高的股票或者其他证券之间出现背离，就应该买进表现相对较差的，卖出表现相对较好的。当未来两者之间的背离得到纠正，那么可以进行相反的平仓操作来获取利润。 

其理论核心在于：单个股票的价格走势接近于几何布朗运动，但考虑到股票间的相关性，股票多空组合的走势有可能接近于白噪声——这意味着股票多空组合的走势可能有固定的均值和固定的标准差。这样，只要我们找到了这样的股票组合，我们就可以做反转交易了，价差的反转交易，即配对交易。

### 7.1.2 多因子$\alpha$策略

多因子$\alpha$策略是找出多个对股票有预测性的因子，用它们来对股票进行预测，得出预测收益率。然后将这些预测组合在一起，形成一个对股票的综合预测，以此来选股。

由此，我们可以看出，任何大尺度的股票策略，都等价于一个预测因子，可看成多因子$\alpha$策略中的一部分。所以，我们说中低频的股票量化就是指多因子$\alpha$策略。

### 7.1.3 股票高频

前面讲到，任何大尺度的股票策略，都可以归入多因子$\alpha$策略。那么小尺度的股票策略呢？它们都可以归入高频策略。高频策略的逻辑都相似，而与非高频策略显著不同，所以我们把高频策略放在一起单列一类来讨论。这样，中低频的股票量化就等价于多因子$\alpha$策略了，在习惯上，我们就以股票量化代指多因子$\alpha$策略。

## 7.2 CTA

CTA是指期货中低频策略。那么疑问就来了，高频呢？为什么是期货而不是现货？CTA是商品交易顾问的简称，而期货并不仅仅包括商品期货呀？

### 7.2.1 为什么是期货而不是现货？

因为重要的大宗商品都是有期货的，期货是标准品，在交易所场内交易，十分方便。而且，期货自带杠杆，可以大幅减小资金占用成本。因此，如果要投资大宗商品，那么，最好的工具就是商品期货。

### 7.2.2 为什么金融期货也在CTA的范畴内？

这是历史造成的。历史上，CTA诞生时，也是以交易商品期货为主。彼时，交易策略就是趋势跟踪。后来，人们发现，指数期货同样具有趋势性，而且，当时，布雷顿森林体系刚刚崩溃，各货币的定价机制尚不完善，外汇期货也具有趋势性。所以，相同的策略可以用于金融期货，人们也就在习惯上把金融期货也纳入CTA的范畴了。

尽管，后来时代变了，外汇期货和利率期货的趋势性变弱了，甚至体现出反转的特性，但其它商品也未尝不是这样。所以，CTA策略逐渐摆脱趋势交易的桎梏，走向了多策略的混合。这样，CTA策略还是同样适用于金融期货，人们也就延续了过去的习惯，把金融期货纳入CTA的范畴。

### 7.2.3 期货高频交易呢？

与前面讨论的股票高频一样，高频策略的逻辑都是相似的，而与非高频策略显著不同，所以我们把高频策略放在一起单列一类来讨论。这样，CTA就仅指期货中低频策略。

## 7.3 高频交易

高频交易是指频率较高的策略，一般平均持仓时间在一分钟左右。高频策略包含多种完全不同逻辑的策略，也应用于不同的交易标的，但同一种策略在不同标的上却是差不多的。

另外，高频策略都对速度要求比较高，对政策环境稳定性要求都比较高，对低廉的手续费要求都比较高，并最终因为这些原因，高频策略最终都将走向赢者通吃的局面。从这些意义上来说，所有高频都是相似的。

高频策略以其超低风险超高收益，深受人们的喜爱与向往，我们后面也会深入讲解。但读者应记住，任何策略都有其弱点，如果你觉得它没有弱点，那一定是它隐藏着非常致命而未被发现的弱点，这是亘古不变的真理。

## 7.4 套利

套利的逻辑和我们前面讲的配对交易相似，我们就不再赘述。下面我们分类讲解。

### 7.4.1 跨市场套利

跨市套利，是指在某个市场买入（或者卖出）某一标的的同时，在另一个市场上卖出（或者买入）同种标的，以期利用两个市场的价差变动来获利。

这个策略主要应用于国际性强的标的，比如，跨国公司的股票、贵金属、有色金属、大豆、原油等。策略的逻辑都相同，也比较简单。于是，大家竞争的焦点就集中在了交易速度上，这最终将导致狠砸IT设备而赢者通吃的局面。

### 7.4.2 跨期套利

跨期套利，是指同一个市场中，不同月份的期货合约应具有某种相关关系，当价差偏离过大，就出现了套利机会。

这个策略主要应用于金融期货、工业品期货这类季节性不明显的标的，大家比拼的是交易速度，最终走向赢者通吃的局面。而农产品期货、能源期货，因季节性明显，跨期套利虽然也可用，但需要考虑的因素变得很多，也更复杂，大家比拼的是对因素考虑的全面程度，所以最终不一定走向赢者通吃的局面。

### 7.4.3 跨品种套利

跨品种套利，是指相近的标的之间，如相互为替代品，其价格应有相关关系，当价差偏离过大，就会出现套利机会。

这个策略常用于三种食用油之间的套利，两种大豆之间的套利，大豆豆油豆粕产业链套利，石油化工产业链套利。这类套利属于风险套利，受多种因素的影响，甚至有时会出现价差走向趋势的情形，大家比拼的是对因素考虑的全面程度，因此，最终也不一定会走向赢者通吃的局面。

### 7.4.4 期权套利

期权套利分风险套利和无风险套利两种。

无风险套利是基于平价公式的。很明显，这种套利也是越套越少，最后拼速度的，所以，最终也会走向赢着通吃。

风险套利的话，又有两种。一种是需要对行情有自己的判断，以此来搭配期权合约。这种拼的是大家对行情的判断，所以，不会走向赢者通吃的局面。另一种是根据隐含波动率曲面来套利，这个有一定风险，所以不见得走向赢者通吃，但门槛是有的，也很高。
