# 第十章 高频策略

## 10.1 导言

### 10.1.1 不可能三角

在投资界，我们常常说存在一个不可能三角——高收益、低风险、大容量。前面介绍的股票量化，是大容量策略，可以做成高收益高风险，也可以做成低收益低风险，但无法做成高收益低风险。而CTA策略，也是大容量策略，一般做成高收益高风险，也有做成低收益低风险的，也无法做成高收益低风险。而本章介绍的高频策略，则可以做成高收益低风险，但代价是无法做到大容量。

### 10.1.2 神秘的高频策略

高频策略在世人眼中，往往是非常神秘的。坊间流传着各种传言，某人特别厉害，用了什么神秘的方法，在短时间内赚了多少倍。无论真假，我们需要记住：

樱桃好吃树难栽，
不下功夫花不开。
幸福不会从天降，
美好生活等不来。

但更要知道，在投资领域，尤其是量化投资，更尤其是高频领域，下了功夫，花还是大概率不会开。下面，我们来讲解为什么会这样，同时解开高频策略的神秘面纱。

## 10.2 高频策略分类

高频策略主要分两类：旨在降低交易成本类型、旨在盈利类型。

### 10.2.1 降低成本类

这类策略，是其它策略的辅助策略。在其它策略需要下单时，它通过对高频数据的分析，得出最优的下单策略，使得平均成交价最优。

#### 10.2.1.1 TWAP

TWAP(Time Weighted Average Price)，时间加权平均价格算法，是一种最简单的传统算法交易策略，主要适用于流动性较好的市场和订单规模较小的交易。

该模型将交易时间进行均匀分割，并在每个分割节点上等量拆分订单进行提交。

例如，可以将某个交易日的交易时间平均分为N段，TWAP策略会将该交易日需要执行的订单均匀分配在这N个时间段上去执行，从而使得交易均价跟踪TWAP。

#### 10.2.1.2 VWAP

VWAP(Volume Weighted Average Price)，成交量加权平均价格算法，是目前市场上最为流行的算法交易策略之一，也是很多其他算法交易模型的原型。

该模型是将一段时间内证券价格按成交量加权得出的平均值，即VWAP是对一段时间市场上所有交易活动平均价格的衡量。

#### 10.2.1.3 PoV

PoV(Percent of Volume)，比例成交算法，同样也是市场上广泛流行的交易算法之一，是VWAP模型的一个改进，算法天然具有反馈修正的能力。

该模型使用实际成交量作为指标，因此在交易时段内总是按照市场成交量的一定比例交易剩余的头寸。

#### 10.2.1.4 冰山算法

这里先要对不熟悉背景的同学介绍一下什么叫做订单簿（Order Book）。现在主流的交易所一般都使用订单簿进行交易，交易所在内部的订单簿上记录所有买家和卖家的报价。

需要强调的是，虽然真正的订单簿只存在于交易所内部，所有交易都在交易所内完成，但是交易所会把每笔报价和市价单都转发给所有人。人们通过跟踪分析自己手里这份的镜像变化，来制定交易策略。

这样带来一个严重的后果，一旦有某个人想要大量买卖，他所发出的巨额限价单会直接展示给所有人。这对他非常不利，因为所有人都会利用这个信息来跟他做对。

为了解决这个问题，有的交易所提供了一种针对性的工具，就是所谓的冰山订单（Iceberg Order）。这种订单可以很大，但只有一小部分是公开出来的，大部分则隐藏起来，除了交易所和发单者本人谁也看不到，真的像一座“冰山”一样。

如果交易所没有提供冰山订单，我们可以通过设计下单策略来模拟实现冰山订单，这种策略就被称为冰山算法。

### 10.2.2 盈利类

这类策略可单独存在，不是其它策略的辅助。它存在的意义就是为了赚钱。

#### 10.2.2.1 Penny Jump

今天假设有一个笨笨的大型机构投资人，他想要买进一只股票，但又不想挂市价买进，所以就在市场里面挂了一张要买进的大单。这时候所有市场里面的人都会看的到订单簿里面有人挂进了大单准备要买进这只股票。

假设市场本来的订单簿是200手10.01的买单和200手10.03的卖单，然后突然这个笨笨的机构投资人进来挂了一张3000手10.01的买单，这时候订单簿会变成3200手10.01的买单和200手10.03的卖单。而我们通常称这个笨笨的机构投资人为"大象"。而高频交易者知道10.01的价位有支撑的买单，所以便会把他们的买价提高1分钱到10.02，而这样的策略就叫做Penny Jump。因为高频交易者知道往下一档的地方，有一只"大象"在支撑着。所以如果价格往上涨到10.03-10.05的话，他就可以马上赚取0.01-0.03的利润。

如果高频交易商买进这只股票之后，就算价格没有往上涨的话，因为下面有一只大象在支撑着，所以他也可以很快的反手用10.01的价位卖给这只大象。

#### 10.2.2.2 Push the Elephant

接着前面那个例子，如果这只大象的确是真的想要买进这支股票，他大概率回将买价提高到10.02。如果观察到这点，高频交易商就不会只满足于只赚每股1分钱的利润，这时候他就可以慢慢的来宰杀这只大象了。

高频交易商的做法是把市场价格推升到10.02，然后等等看，看这只大象会不会把他们买进价格也跟着提升到10.02，如果大象买进价也跟着提升到10.02的话，代表着这个大象很有可能已经掉入了高频交易商的陷阱了。

接着高频交易商会再进一步的把价格往上推升，推升到10.03，重复看大象的买进价有没有跟上来，如果有的话就再继续往上推。假设最后推到10.05，这时候大象的买进价已经被从10.01拉高到了10.05，那么这时候高频交易商就可以把所有的库存全部倒给这只大象，来赚取利润。

因为这样的策略是希望一路吸引着大象，把他们的买进价格1分钱,1分钱的慢慢推高，所以也称之"Push the elephant"。

#### 10.2.2.3 Market Maker

做市商(market maker)为证券交易所指定的买卖中间商，主要业务是为买方及卖方进行报价、并且为双方寻找好的价格撮合交易，本身则从买卖差价中获利。

在大多数电子化的连续交易市场中，即使我们不是做市商，我们也可以模仿做市商的报价行为，赚取相应的利润。

做市商对市场起到很好的作用：
1. 提高股票的流动性，增强市场对投资者和证券公司的吸引力。
2. 有效稳定市场，促进市场平衡运行。
3. 具有价格发现的功能。

#### 10.2.2.4 Take Out Slow Movers

在某些时间点，市场价格可能会快速且大幅度的变动，比如说经济数据报告公布或某公司有重大利多/利空消息公布的时候。那么速度比较快的交易商便可以很快的对这种价格变动做出反应，而那些反应比较慢的交易商，自然就成了速度快交易商的午餐了。

这类策略主要抓两种机会：
1. 事件性机会，比谁反应快。
2. 市场中有傻子报出了不合理价格的订单，比谁先把傻子的订单吃掉。

所以对专门作高频交易的人来说，为什么网络/计算机/软件/算法的速度那么重要的原因是如此，在高频交易的世界里，只有速度最快的人才能赚到钱。

#### 10.2.2.5 Scratch for the Rebate

在美国的股票市场里，交易所有时候会为了提高股票的流动性，而对做市商所报价而成交的股票有奖励机制，这也就是一般所俗称的Rebate。

这样，很多做市商在交易时，就不求赚钱了，10块买10块卖，在市场中一分钱不赚，转而去赚交易所的奖励。这就叫Scratch for the Rebate。


## 10.3 国内高频策略现状

国内的高频交易和国外差距不大。在国内从事高频交易的人，大部分也是从国外回来的，所以总体差距较小。下面我们讲讲在全市场中影响较大的几个策略。

### 10.3.1 第一波2011年——股指期货跨期套利

2011年，股指期货上市初期，参与者不够成熟，市场上有大量的套利机会。其中，以跨期套利最为著名。跨期套利的核心在于不同期货合约之间的价差，在短期内应保持比较稳定的状态。而现实中，这个价差在2011年经常大幅波动，这就给了套利者无穷的机会。在那个时代，年收益可以做到30%以上。

然而，随着市场参与者越来越成熟，到2013年，跨期套利的机会已经大幅减少了。尤其到了2014年底，这个价差甚至在短期内出现趋势性特征。这给跨期套利策略带来了灾难。到2016年，本策略的竞争越来越激励，慢慢就进入了赢者通吃的状态——除了个别实力极强者之外，其它参与者逐步退出市场。

### 10.3.2  第二波2015年——股指期货超高频

在2015年以前，股指期货的超高频就存在，只是容量及其有限，没有在市场上掀起什么风浪。然而到了2015年，由于大量散户在赚钱效应下入市，市场非理性程度大幅提高。这给了股指期货超高频交易非常大的机会。这里，我们说的超高频策略指平均持仓时间在5秒之内的策略。

尤其到了4月，IH和IC上市后，大量跨品种套利者涌入，给股指期货的超高频带来了更多机会。6月股灾开始后，本策略迎来了前无古人后无来者的盛筵，短短一两个月，10倍20倍且无回撤者，比比皆是。

但是，高收益低风险必然低容量，如果容量还不低，那等待我们的将是游戏规则的修改。9月，中金所修改股指期货开仓限制以后，本类策略全部无法继续，盛筵到此为止。

### 10.3.3  第三波2019年——股票T+0

2018年，我们经历了一整年的大熊市。2019年初，监管层放松了股票短线交易的监管，市场热情迅速被点燃。各种手工的、程序化的股票T+0策略层出不穷。这类策略在2019年的前5个月获得了相当不错的收益，优秀者能做到年化20%-30%，最大回撤1%以内。其中，激进者可通过券商加约4倍杠杆，获得年化80%左右的收益。

这类策略的平均持仓时间在1分钟左右，持仓时间长者可达1小时，短则数秒钟。其策略以事件性的短期趋势为主，兼以超跌反弹为辅。而且，因股票类策略本身容量就大于期货策略，一时间，受到大量资金的追捧，洛阳纸贵。据行业人士估计，短时间内涌入该策略的资金约1000亿。

自6月起，一方面竞争日趋激烈，另一方面市场波动下降，策略逐步失效。上半年加了杠杆者，悉数陷入窘境，未加杠杆者也只能艰难维持。至12月，市场波动率上升，才得到缓解。但无论如何，赚大钱的日子已经一去不复返了。
