# 第一章 证券投资历史

## 1.1 史前时代——证券市场出现以前

### 1.1.1 人猿相揖别

自人之为人之时，人就是群居动物，会合作猎杀大型动物，比如猛犸。这就是众筹。

### 1.1.2 铜铁炉中翻火焰

濮阳人吕不韦贾于邯郸，见秦质子异人，归而谓父曰：“耕田之利几倍?”曰：“十倍。”“珠玉之赢几倍?”曰：“百倍。”“立国家之主赢几倍?”曰：“无数。”曰：“今力田疾作，不得暖衣余食;今建国立君，泽可以遗世。愿往事之。”

### 1.1.3 几千寒热

高利贷，即债权投融资，自古有之，绵延数千年。

## 1.2 萌芽时期——证券市场初期

### 1.2.1 荷兰

16世纪末，东西方之间的香料贸易盛行，欧洲各国纷纷成立航运公司。渐渐地，竞争逐步激烈起来，由于国力不足，荷兰在与英国、西班牙的竞争中处于劣势。荷兰就想到了成立国家级联合公司，来参与竞争。那么，为平衡各方利益，公司采用股份制，荷兰全民可参与。

于是，1602年，荷兰东印度公司成立，印制了世界上最早地股票。由于全民持股，渐渐地，荷兰人民交易东印度公司股票地需求越来越强烈。于是，1611年,世界上最早地交易所——阿姆斯特丹证券交易所开始使用。

不过，区区几十年之后，荷兰就发生了郁金香泡沫。历史真是有趣。

### 1.2.2 英国

英国也是资本市场发展较早的国家之一。1711年，英国南沙公司成立。其本质是一个政府融资平台，用于偿还战争时期的债务。同时，政府授予它在拉美地区的贸易垄断权，以此获取收益。1720年，南沙公司贿赂政府，使国会通过了南沙股票换国债计划，使南沙股票大受追捧。很快南沙股票从120英镑涨到了1000英镑以上，不久又跌倒了190英镑。

牛顿就参与了其中，他在100+英镑时买入，300+英镑时卖出，然后价格到500+时，他看到朋友赚了大把的钱，一番纠结之后，终于在800+时大举买入，最后在200+时清仓割肉。

### 1.2.3 美国

美国证券市场萌芽于独立战争时期，即南沙泡沫的几十年之后。经过了100+年的发展，到1929年大萧条以后，美国政府加强了对证券市场的立法和控制，才逐步规范化。

## 1.3 成熟时期——风云人物

经过了罗斯福新政和二战以后，到1954年，人们终于摆脱了大萧条以来对股市的阴影。股市逐步活跃起来。

### 1.3.1 对冲基金初创期

#### 1.3.1.1 阿尔弗雷德·温斯洛·琼斯

琼斯可谓是对冲基金的鼻祖。他于1949年创立了史上第一只对冲基金，同时做多潜力股和做空垃圾股。至1968年，他的累计回报率达到5000%，年化22%的收益。他开创了2%+20%的收费方式，被行业沿用至今。他还发明了“多经理人竞争”模式，筛选出了优秀的选股经理人。

可是，成也萧何，败也萧何。在牛市中，他逐步抛弃了对冲的思想，认为“做空是失败者才会去做的事”。同时，牛市中，“多经理人竞争模式”导致谨慎的投资经理在竞争中逐步被淘汰。到了1969年，牛市终结了，琼斯亏损了35%，比大盘亏得还多。他的传奇也终结了。              

### 1.3.2 承前启后

#### 1.3.2.1 迈克尔·斯坦哈特

斯坦哈特是个特立独行的人。他承上启下，活跃在20世纪70年代和80年代。

他在1967年创立了自己的基金。此时，正值牛市的高潮，也许是性格原因，他选择了与众不同，他逐步对冲了他的持仓，甚至，还逐步增加了空头的暴露。当1969年股价下挫时，他逃过了亏损。而70年代初，股价进一步下挫时，他甚至还获得了盈利。

他聘请投资经理也遵循特立独行的原则。他的投资经理弗兰克·席鲁夫开创了以利率预测股价的理论，并成功应用于实践，在70年代的滞胀中，赚取了不小的利润。

赚到钱之后，他开始转向场外的大宗交易，成为了场外大宗交易市场的做市商，赚的盆满钵满。然而，这也引起了监管层的注意。80年代后期，监管加强，场外大宗交易市场萎缩，他也逐渐消失在历史中了。

#### 1.3.2.2 海默·威玛

威玛是萨缪尔森的学生。1970年，他创立了自己的基金——商品公司。期初，他用学生时代的计量经济学模型，对商品的基本面建模，依据模型给出的估价来买卖商品期货。然而，仅仅一年之后，他就失败了——他的模型预测玉米会下跌，他就做空，尽管玉米处于不断向上的趋势，最终招致惨败。威玛很快从失败中总结了教训——风控止损必须一开始就编入计算机程序。

1974年，迈克尔·马库斯加入了商品公司。马库斯开创了一种新的交易策略——趋势跟踪。这与威玛过去的模型格格不入。然而，马库斯不断地盈利让大家都闭嘴了。由此，商品的趋势交易逐步风靡全球。直至今天，这个策略仍然在盛行，它被称为CTA策略。

70年代末，马库斯逐步把关注点移到货币交易上。他通过发掘汇率政策有错误的国家，通过做多或做空该国货币，获取利润。他应该算是索罗斯的前辈了。

1977年，马库斯招聘了一名助理——布鲁斯·科夫勒。不久，科夫勒发现了远期汇率的定价错误——人们往往只用汇率来给远期汇率定价，而忽视了利率和国际资本流动的影响。从而，科夫勒开创了连续10年平均80%的收益。

然而，科夫勒的成功终结了商品公司——当客户纷纷要求把资金全部交给科夫勒管理时，科夫勒离开单干了。但无论如何，商品公司是多个领域的开创者，为后来的商品类策略奠定了基础。

#### 1.3.2.3 朱利安·罗伯逊

罗伯逊在1980年创立了老虎基金。他本人是个选股高手。他也招了一大批选股高手作为助手。他的助手们不断选出优秀的股票，经过他的评判，赋予权重，进行交易。优秀的助手将获得更大的权重，差的助手将被解雇。这是一种新的模式，对后来的量化投资有很大的启发。

然而，这种模式也带来弊端——优秀的助手纷纷脱离公司独立创办基金，于是到了90年代，老虎基金也逐渐开始走下坡路了。

### 1.3.3 百花齐放

#### 1.3.3.1 乔治·索罗斯

索罗斯在1969年创立双鹰基金，1978年改为量子基金。80年代，布雷顿森林体系崩溃后，汇率出现大幅波动，这给了他机会。他将精力放在货币交易上，寻找定价错误的货币，做多或做空，甚至操纵市场，并以此成名。

1988年，斯坦利·德鲁肯米勒加入量子基金。几年后，他发现了欧洲汇率市场的不平衡。于是，在索罗斯的支持下，他完成了英镑狙击战，一战成名。

1993年，阿米尼奥·弗拉加加入了索罗斯。几年后，他发现了泰国经济的问题。于是，德鲁肯米勒和索罗斯迅速行动，完成了泰铢狙击战。这导致了东南亚经济危机。

然而，索罗斯不仅仅是个投资家，还是个慈善家。他希望能通过投资，暴露出对手的问题，迫使对手做出改变，走上正确的道路。这种双重性格使这次交易并未带来太多利润。更令人哭笑不得的是，在泰铢狙击战的同时，他还同时狙击了俄罗斯，并失败了。这次失败的原因正是他希望帮助俄罗斯转轨，从而一边打一边帮。

虽然试图去改变他国的行为很可笑，但无论如何，他是个投资天才。后人将他的这种策略称为宏观对冲策略。

#### 1.3.3.2 保罗·都铎·琼斯

相对于三巨头斯坦哈特、罗伯逊、索罗斯，都铎·琼斯则是资本市场的新秀。他于1983年创立自己的基金，以投资商品为主。他非常善于去影响其他投资者的心理，创造出一段趋势，并从中获利。以今天的观点来看，这显然是操纵市场。但在80年代末90年代初，他是风云人物。

#### 1.3.3.3 约翰·梅里韦瑟

梅里韦瑟这个名字，可能对大家比较陌生。但他1994年创立的长期资本却是家喻户晓。他们的特色是套利。然而，随着竞争者越来越多，套利空间越来越小，他们不得不使用杠杆来放大收益。虽然在最初的几年里，他们收益可观，但套利交易小赚大亏的特性，使得杠杆交易潜藏着巨大的风险。终于，在索罗斯狙击东南亚和俄罗斯之时，小概率事件发生了，他们的套利组合发生了巨大的亏损，爆仓了。

#### 1.3.3.4 詹姆斯·西蒙斯

西蒙斯，这位数学家应该是量化投资界最著名的人物了。1988年，他成立了大奖章基金，于是，传奇开始了。30年来，复合收益率超过30%。而这还是他扣除了5%的管理费和44%的业绩报酬之后的数字。

要说策略，他一直很神秘。但从近期流传出来的资料看，似乎有没有太特别的东西，只是走在了竞争者的前面而已。但无论如何，他们的东西偏高频，这是无疑的。毕竟他们的核心策略的容量不足百亿美元，只供内部人投资，而对外发售的基金，却表现很差。这大概率是因为内部人投的是高频策略，外部人投的是低频策略。

### 1.3.5 阿尔法工厂

90年代，理查德·格林诺德和雷诺德·卡恩将自己在量化投资领域的研究写成了一本书《主动投资组合管理》。这本书迅速风靡全球。它被成为量化投资的“圣经”。这本书告诉我们，如果我们能找到一些有预测力的因子，那么，我们怎样最优地利用它们，以形成我们的持仓。作者给出了数学证明，证明了这种方式是最优的。由此，量化投资变得容易了，对冲基金只需要招一群人来寻找阿尔法（有预测力的因子），然后按照书上给出的方法形成持仓，就完成了量化投资。

进入21世纪以后，这种模式也迅速风靡全球的对冲基金，成为量化投资的标准模式——阿尔法工厂。

## 1.4 国内——跨越式发展

中国的一天，相当于美国的一年。虽然夸张了，但中国的证券市场，确实用了区区30年的时间，走完了美国近200年的路。而量化交易，也用了10年，走完了美国30年的路。

我们重点讲讲我国量化投资的发展历程。

## 1.4.1 试水2009——指数增强拉开帷幕

2008年金融危机之后，大量国外人才回到国内，推动了国内量化投资的发展。2009年，李笑薇加盟富国基金，同年发行了富国沪深300指数增强基金，我国量化投资正式拉开帷幕。

## 1.4.2 滑稽2010——股指期货上市

转眼到了2010年4月，股指期货IF上市。资本市场从此有了对冲工具，为对冲策略开辟了发展的道路。然而，对冲的基础在于期货在交割时，终将收敛于现货。在2010年的大部分时候，期货竟不收敛于现货，于是乎，一批做期现套利的人，发了财。有的狠人，甚至取得了百倍的收益。

## 1.4.3 启程2011——程序化交易元年

2011，期货交易所放开了程序化交易接口，从此，任何一个会写程序的人，都可以用程序来交易期货了。期货市场得到了巨大的发展。同时，期货逐步开始收敛到现货。各种对冲策略逐步成熟，股票量化对冲基金得到了飞速发展。

期货量化形成了以趋势交易为主，套利、高频为辅的业态。其中，趋势交易，赚了大钱；期货套利交易，也赚了不少；期货高频交易刚起步，也赚了不少。

股票量化，则形成了以市值因子为主的单因子策略。行业普遍号称是多因子，但实际市值因子就占了50%以上的权重。大家纷纷做多小市值股票，做空期货。这在2011年，也是赚了不少的。

## 1.4.4 转折2014——股票量化的灾难

2014年期货量化平稳发展，在经历了2013年的短暂沉寂之后，趋势策略在2014年取得了大发展，翻倍策略比比皆是。

2014年第四季度，由于大市值股票大涨，指数也跟随大涨，但小市值股票趴着不动。市值因子反转，导致重仓市值因子的基金一个月亏掉了过去三年的盈利。市场一片哀嚎。

这其实是一次产能出清，股票量化就此沉寂，直到2018年。

## 1.4.5 疯狂2015——期货量化盛极而衰

继2014年期货趋势策略表现抢眼之后，2015年，期货量化全面爆发。趋势策略继续20%-30%的涨幅，套利策略也盆满钵满，高频策略更是取得了10-20倍的收益。

但也因此，引起了监管层的注意。股指期货全面封杀，商品期货手续费大幅提高。期货量化盛极而衰。

## 1.4.6 债券2016——期货量化的回光返照

2016年，债券出现了大牛市。而当时的监管允许国债无限放杠杆，导致固定收益类策略竟获得了40%以上的收益。

同时，伴随着去产能的大政策，黑色金属迎来了巨大波动。相应地，期货趋势交易也赚了大钱。期货趋势类策略，由于自带杠杆，为控制风险，往往有7成资金放在固定收益类策略上。于是强强叠加，使得期货趋势类基金普遍获得70%左右的收益。

超高的收益使得巨量的资金涌入了期货这个本不宽敞的市场。期货交易所为了控制风险，不断提高手续费标准。债券的监管也禁止了无限放杠杆的行为。这使得2016年，成了期货量化最后的狂欢。自此以后，期货量化陷入低迷。

## 1.4.7 冰火2018——股票量化的第二春

经历了2017大盘股的牛市以后，2018年，股市又吸引了不少的资金。然而，自中美贸易战起，一整年的大幅下跌，让市场陷入寒冬。市场信心跌落谷底。

与此同时，股票量化对冲策略异军突起。该类基金都已用上了国际上最先进的阿尔法工厂模式，在2018年股市下跌30%的情况下，取得了20%-30%的稳定收益，收益回撤比纷纷达到10左右。

即使是股票量化指数增强策略，也仅仅亏损了不到10%，相比于市场，大幅跑赢。

这样的业绩对比使整个市场大为震惊，纷纷感叹人工智能的崛起和量化投资的魔力。


## 1.4.8 末路2019——股票量化盛极而衰

经历了2018的大熊市之后，2019年第一季度，股市大幅反弹。在高波动率的刺激下，股票量化迭创新高。股票量化对冲策略，一个季度赚10%的比比皆是。而股票量化指数增强策略，更是在一季度赚到了40%-50%。

如此傲人的业绩吸引了巨量的资金。甚至银行理财资金，也以配资的形式进入股票量化投资。行业规模在短短半年内，从500亿迅速增长到1500亿。然而，万事万物，盛极而衰。对于任何超过合理收益的投资，当你看不到它的风险在哪里的时候，就是最危险的时候。

到了6月以后，股票量化对冲策略持续亏损，股票量化指数增强策略相对于指数的超额收益也增长缓慢。股票量化行业陷入无助，阿尔法工厂们不断对预测因子进行更新迭代，但收效甚微。到了12月，经过半年的无所作为，投资人纷纷赎回，行业遭受了巨大冲击。

未来路在何方？谁也说不清。
