# 第二章 证券投资理论

## 2.1 市场的有效性

1970年，尤金·法玛提出并深化了有效市场假说。

假说认为，在法律健全、功能良好、透明度高、竞争充分的股票市场，一切有价值的信息已经及时、准确、充分地反映在股价走势当中，其中包括企业当前和未来的价值，除非存在市场操纵，否则投资者不可能通过分析以往价格获得高于市场平均水平的超额利润。

显然，其假设并不成立。于是，他又放宽了其假设，提出了有效市场的三种形式：

### 2.1.1 弱式有效市场

该假说认为在弱式有效的情况下，市场价格已充分反映出所有过去历史的证券价格信息，包括股票的成交价、成交量，卖空金额、融资金额等。
如果弱式有效市场假说成立，则股票价格的技术面分析失去作用，基本面分析还可能帮助投资者获得超额利润。

### 2.1.2 半强式有效市场

该假说认为价格已充分反映出所有已公开的有关公司营运前景的信息。这些信息有成交价、成交量、盈利资料、盈利预测值、公司管理状况及其它公开披露的财务信息等。假如投资者能迅速获得这些信息，股价应迅速作出反应。
如果半强式有效假说成立，在市场中利用基本面分析则失去作用，内幕消息可能获得超额利润。

### 2.1.3 强式有效市场

强式有效市场假说认为价格已充分地反映了所有关于公司营运的信息，这些信息包括已公开的或内部未公开的信息。
在强式有效市场中，没有任何方法能帮助投资者获得超额利润，即使基金和有内幕消息者也一样。

## 2.2 马科维茨——投资组合理论

### 2.2.1 投资组合理论的主要内容

1952年，马科维茨首次提出了投资组合理论。该理论包含两个重要内容：均值-方差分析方法和投资组合有效边界模型。

人们进行投资，本质上是在不确定性的收益和风险中进行选择。投资组合理论用均值-方差来刻画这两个关键因素。所谓均值，是指投资组合的期望收益率，它是单只证券的期望收益率的加权平均，权重为相应的投资比例。所谓方差，是指投资组合的收益率的方差。我们把收益率的标准差称为波动率，它刻画了投资组合的风险。

人们在证券投资决策中应该怎样选择收益和风险的组合呢？这正是投资组合理论研究的中心问题。投资组合理论研究“理性投资者”如何选择优化投资组合。所谓理性投资者，是指这样的投资者：他们在给定风险水平下对期望收益进行最大化，或者在给定期望收益水平下对风险进行最小化。

因此把上述优化投资组合在以标准差为横坐标，期望收益率为纵坐标的二维平面中描绘出来，形成一条曲线。这条曲线上有一个点，其标准差最低，称之为最小方差点，对应的组合被称为最小风险组合。这条曲线在最小方差点以上的部分就是投资组合有效边界，有效边界上投资组合称为有效投资组合。

在标准差-期望收益率二维平面上，任意一个投资组合要么落在有效边界上，要么处于有效边界之下。因此，有效边界包含了全部（帕雷托）最优投资组合，理性投资者只需在有效边界上选择投资组合。

### 2.2.2 投资组合理论的意义

投资组合理论首次对风险和收益这两个投资管理中的基础性概念进行了准确的定义，从此，同时考虑风险和收益就作为描述合理投资目标缺一不可的两个要件。

投资组合理论关于分散投资的合理性的阐述为基金管理行业的存在提供了重要的理论依据。

“有效投资组合”的概念，使基金经理从过去一直关注于对单个证券的分析转向了对构建有效投资组合的重视。

投资组合理论虽未直接应用于股票投资，但已被广泛应用到了大类资产配置中，并被实践证明是行之有效的。

### 2.2.3 投资组合理论的不足

投资组合理论需要计算所有投资标的两两之间的协方差。在股票市场中，由于标的太多，计算量太大。这种协方差的计算几乎是不可能的。也正因为此，它在实践中仅被应用于大类资产配置。

在股票市场中，我们对期望收益、协方差的估计误差较大，这种误差带来的影响也非常大，从而给解带来很大的不稳定性。任何一点小小的误差，都会带来持仓的大幅不同，从而产生较大的调仓成本。很多时候，我们会发现，我们优化的对象不是收益，而是误差。


## 2.3 夏普——单指数模型

夏普是马科维茨的学生，他在1963年发明了单指数模型，成功解决了马科维茨投资组合理论中的一个难题——难以估计的协方差。

他定义了一个“市场组合”，假设股票间两两的相关性是由股票与市场组合的相关性带来的。这样，我们只需要计算股票与市场组合的协方差，就可以推算出股票两两间的协方差了。

单指数模型的出现大大降低了马科维茨投资组合理论的计算量。但计算误差的问题，仍未解决。

## 2.4 CAPM——资本资产定价模型

1964年，刚发明了单指数模型的夏普，突然转念一想，发现单指数模型竟然可以用于给股票定价！于是就在此基础上发明了资本资产定价模型（CAPM）。

CAPM假设所有投资者都按马科维茨的资产选择理论进行投资，对期望收益、方差和协方差等的估计完全相同，投资人可以自由借贷。

CAPM定义了资产 $i$ 的超额收益为 $R_i - R_f$ ，市场 $m$ 的超额收益为 $R_m - R_f$ ，其中 $R_f$ 是无风险收益。然后，定义了
$$\beta_i = \frac{\mathbf{Cov}[R_i - R_f, R_m - R_f]}{\mathbf{Var}[R_m - R_f]}$$

然后，CAPM说明，对于不同股票而言，其超额收益与其 $\beta$ 成正比。即超额收益反映了股票的 $\beta$ ,即反映了股票的市场风险，而不反映其非市场风险。换言之，股票的超额收益是其市场风险的补偿。即
$$\mathbf{E}R_i - R_f = \beta (\mathbf{E}R_m - R_f)$$

## 2.5 APT——套利定价理论

12年之后，1976年，斯蒂芬·罗斯在CAPM的基础上提出了APT（套利定价理论）。APT使用了比CAPM更简单的假设，尤其是它没有假定“市场组合”的存在性。同时，APT又得出了更漂亮的结果——即股票的超额收益与它在一系列风险因子上的暴露成正比，而这些风险因子的因子收益率不因股票的不同而不同。即
$$\mathbf{E}R_i - R_f = \sum_j \beta_{ij}\mathbf{E} R_j$$

## 2.6 当代理论介绍

前述都是现代证券投资理论，是经典理论。而80年代以后，逐步出现了一些新理论，它们也有突出的贡献，但尚未盖棺论定，所以我称之为当代理论。经典理论虽然很漂亮，但实用性不强，它们更多地还是在帮助我们打好基础；当代理论实用性强，则更偏重于指导我们地投资实践。我们在这里简要介绍一下：

### 2.6.1 Fama-French三因子模型

在CAPM的基础上，1992年，Fama和French将定价模型扩展为三因子。除了CAPM中的市场因子，他们还引入了市值因子和成长因子，形成了三因子模型。模型认为，股票的超额收益是这三个因子的风险补偿的线性组合。

### 2.6.2 Black-Litterman模型

如前所述，马科维茨的投资组合理论存在一些不足。最核心的是三点：期望收益预测误差大、协方差预测误差大、协方差计算量大。如果仅将其应用于大类资产配置，则关于协方差的问题都能得到解决。那么，唯一的问题就是期望收益预测误差大。这会导致计算最优持仓量时，结果波动非常大。

1992年，Black和Litterman使用市值来作为市场对股票的持有量，倒推出市场对股票的预期收益率。然后，他们将个人对股票的预期收益率和市场对股票的预期收益率加权求和，以此为预期收益率，再计算出最优持仓量。这样，经过加权求和以后的预期收益率就会和市场对股票的预期收益率非常接近。从而，最优持仓量也会和市场对股票的持有量非常接近。

这里，Black和Litterman用到了贝叶斯的思路。虽然模型并不美，但确实解决了问题。

### 2.6.3 Barra风险模型

在学术界还在研究三因子模型的时候，业界已经开始把它做成产品拿来卖钱了。是的，一家叫Barra的公司，找了68个因子，做了一个风险模型，用于解释股票的超额收益，也即解释其风险来源。

最基本地，任给一个股票组合，Barra都可以告诉你组合在68个风险因子上的暴露是怎样的。

进一步说，任给一个股票组合，Barra都可以帮你重新配置权重，使得其风险收益特征最优。

最后，任给一个股票组合的历史持仓，Barra都可以帮你评估该基金经理的盈利来源是哪里——是实力强？还是运气好？

### 2.6.4 经典理论的改进——基于主动头寸

进入本世纪以后，残酷的现实让人们发现，绝大部分的基金经理，并不能跑赢指数。由此，被动投资（指数投资）逐步兴起。与此同时，主动投资也在逐步向指数增强策略转型。即，主动投资经理往往会告诉客户：他的业绩比较基准是某某指数，他会在控制与该指数的跟踪误差的同时，争取一些超过指数的收益。

这种变化，使得人们逐步开始改造经典理论。这种改造的核心就是基于主动头寸，用主动收益和主动风险去替代经典模型中的总收益和总风险。事实证明，这种改造是可行的，业界由此创造出了基于主动头寸的新理论，并应用于实践。

同学们不妨想一想，前面的传统理论，可以怎样去改造？

### 2.6.5 主动管理第一定律

在基于主动头寸改造经典理论时，人们将传统的夏普比改造为信息比。
$$SR = \frac{\mathbf{E}[R_p] - R_f}{\mathbf{Std}[R_p]}$$
其中，$SR$是夏普比，$R_p$是组合的年收益，$R_f$是无风险收益。夏普比反映了承担单位总风险而获取的总收益。
$$IR = \frac{\mathbf{E}[R_p - R_b]}{\mathbf{Std}[R_p - R_b]}$$
其中，$IR$是信息比，$R_p$是组合的年收益，$R_b$是业绩比较基准的年收益。信息比反映了承担单位主动风险而获取的主动收益。

经过人们的研究，人们证明了一个非常有意义公式——主动管理第一定律：
$$IR = IC \cdot \sqrt{BR}$$
其中，$IR$是信息比。$IC$是信息系数，等于投资经理预测准确度（预测的主动收益和实际主动收益的相关系数）。$BR$是投资经理平均每年做预测的次数。

从这个定理，我们可以得到如下推论：

择时不如选股。因为增加投资标的可以大幅增加$BR$，这非常容易。而要增加预测准确度$IC$，则非常难。一方面，这需要多年的研究，多年的付出，多年的积淀；另一方面，因为$IC$是相关系数，相关系数本身，是有上限的。


### 2.6.6 信息理论

信息理论的核心在于：投资经理给出了原始预测，我们需要把它转化成预期收益，在这个转化过程中，如何尽可能多得保留投资经理给出的信息？它由多个理论组成，包括：

如果只考虑单期，如何在原始预测的基础上求得预期收益的最小方差无偏估计。

原始预测经过时延之后，是否依然有效？是否会衰减？如何衰减？

在本期求取预测收益时，如何加入上期原始预测的残余影响？上上期的呢？

截面上的原始预测和时间序列上的原始预测在求取预期收益时有何不同？

这些问题很现实，很具体，同学们可以自己思考一下。
