# 第九章 CTA策略

## 9.1 导言

CTA策略与股票量化的区别，不可谓不大。而这种区别，直接影响CTA策略的研发流程。

这种区别的核心是标的不同。股票量化是投资股票，而CTA是投资期货。根据期货品种不同，还可以分为农产品、有色金属、化工品、贵金属、能源、金融等。标的的不同带来波动性质的不同。这意味着如果统一建模，要找到足够的预测因子较难。更恰当的方式是各品种各行业单独建模。而且，单独建模时也应考虑品种间的联动。

另外，大宗商品一般周期性较强。它受大尺度的经济周期影响较大。而且，随着科技的进步，不同品种在产业中的地位也在变化，这导致我们很难使用长期数据来建模，也很难建立较大尺度的模型。

以上两个问题，导致对CTA策略总体而言，有效的数据量偏少。因此，CTA策略普遍尺度偏小，以利用更多的高频数据。CTA策略一般持仓时间在数周以下，绝大部分的持仓时间在数小时至一两日。但即使如此，有效的数据来仍然偏少，这对建模造成了极大的困难。

无论如何，我们先分别了解一些典型的期货品种有何特征。

## 9.2 典型期货品种的特征

### 9.2.1 股指期货

股指期货的标的是股票价格指数。股票价格指数为度量和反映股票市场总体价格水平及其变动趋势而编制的股价统计相对数。通常是报告期的股票平均价格或股票市值与选定的基期股票平均价格或股票市值相比，并将两者的比值乘以基期的指数值，即为该报告期的股票价格指数。当股票价格指数上升时，表明股票的平均价格水平上涨；当股票价格指数下跌时，表明股票的平均价格水平下降；是灵敏反映市场所在国（或地区）社会、政治、经济变化状况的晴雨表。

尽管股指是由个股加权平均而来，但它与个股表现出了显著不同的特性。

经过统计，我们会发现个股的价格呈较为明显的反转特性——即大涨后回落的概率比继续涨的概率大，大跌后反弹的概率比继续跌的概率大。理论上，个股的价格反映上市公司的价值，价格围绕价值波动，那么，它呈现反转特性也算合乎逻辑。

而同时，经过统计，我们也会发现股指呈较为明显的趋势特性——即大涨后继续涨的概率比回落的概率大，大跌后继续跌的概率比反弹的概率大。理论上，股指反映所在国（或地区）的宏观经济，宏观经济的变化具有趋势性，那么，它呈现趋势特性也算合乎逻辑。

但两个统计结果放在一起时，却总让人感到莫名的奇怪。但趋势交易，无疑是股指期货的主角了。

另外，股指受政策影响较大，但不同国家（或地区）的政策往往不同。这跟执政能力、执政经验、文化风格有关。这意味着，我们在做股指期货的趋势交易时，还应考虑如何把政策的因素纳入模型。

再者，不同市场的做空机制不同，涨跌停板/熔断机制不同，这也导致股指的波动性质不同。这些因素也应纳入模型的考量。

### 9.2.2 利率期货

利率期货的标的一般是国债价格。国债又称国家公债，是国家以其信用为基础，按照债券的一般原则，通过向社会发行债券筹集资金所形成的债权债务关系。国债是中央政府为筹集财政资金而发行的一种政府债券，由中央政府向投资者出具的、承诺在一定时期支付利息和到期偿还本金的债权债务凭证，由于国债的发行主体是国家，所以它具有最高的信用度，被公认为是最安全的投资工具。

国债价格高，意味着市场利率低。国债价格低，意味着市场利率高。因此，利率期货的价格反映了市场对于未来市场利率的看法。

对于一个成熟的经济体，利率市场的典型特征就是稳定。那么利率期货无疑是反转交易唱主角了。但是，在不那么成熟的经济体，利率市场波动就大了。甚至，对一些债务危机深重的国家，利率市场随时可能失控，到那时，趋势交易就应该唱主角了。所以，总体说来，利率市场尖峰厚尾的特性尤为明显。而不同的国家（或地区）中，更侧重于尖峰还是厚尾，则依赖于它在国际金融体系中的地位，与内部政策对经济的影响力。

另外，汇率波动也会影响利率。当一国汇率贬值时，如果该国（或地区）不希望汇率贬值，那么它就很难维持较低的利率。

再者，宏观经济也会影响利率。当一国经济过热时，如果该国（或地区）不希望硬着陆，那么，它就很难维持较低的利率。

### 9.2.3 外汇期货

外汇期货的标的是汇率。外汇市场是指在国际间从事外汇买卖，调剂外汇供求的交易场所。它的职能是经营货币商品，即不同国家的货币。
国际上因贸易、投资、旅游等经济往来，总不免产生货币收支关系。但各国货币制度不同，要想在国外支付，必须先以本国货币购买外币；另一方面，从国外收到外币支付凭证也必须兑换成本国货币才能在国内流通。这样就发生了本国货币与外国货币的兑换问题。两国货币的比价称汇价或汇率。西方国家和我国的中央银行为执行外汇政策，影响外汇汇率，经常买卖外汇的机构。所有买卖外汇的商业银行、专营外汇业务的银行、外汇经纪人、进出口商，以及其外汇市场供求者都经营各种现汇交易及期汇交易。这一切外汇业务组成一国的外汇市场。

和利率类似，对于一个成熟的经济体，汇率的典型特征也是稳定。所以外汇期货无疑也是反转交易唱主角。但是，汇率受国际形势、国际贸易的影响非常大。同时，它也受政策影响很大，根据蒙代尔三角理论，每个国际（或地区）都有自己的政策选择，这催生了完全不同的利率、汇率运行特征。因此，在选择趋势交易或反转交易时，要考虑的东西变得更多。

同样，外汇市场也会受利率市场的影响。这种影响也同样复杂。

### 9.2.4 黄金期货

黄金期货的标的是黄金。黄金是贵金属的代表，具有一定的货币属性，因此，是一种避险资产。在和平时期，黄金的货币属性较弱，而在动荡时期，黄金的货币属性较强。

20世纪70年代以前，黄金价格基本由各国政府或中央银行决定，国际上黄金价格比较稳定。70年代初期，黄金价格不再与美元直接挂钩，黄金价格逐渐市场化，影响黄金价格变动的因素日益增多。

作为一种避险资产，黄金的短期波动性质就比较接近于带跳的几何布朗运动。相对来说，它的波动较大部分来源于全球重大事件的影响。因此，黄金的小尺度策略更倾向于跨市场套利。

另外，由于黄金具有一定货币属性，是各国（或地区）的储备资产。因此，全球大环境的变化对各央行的行动产生影响，从而带来大尺度的趋势。所以，黄金的大尺度策略是以趋势交易为主的。

### 9.2.5 原油期货

原油期货的标的是原油。原油是能源的代表，也具有一定的货币属性。但是，由于其工业用途也非常广泛，因此，不算是一种避险资产。

习惯上把未经加工处理的石油称为原油。一种黑褐色并带有绿色荧光，具有特殊气味的粘稠性油状液体。是烷烃、环烷烃、 芳香烃和烯烃等多种液态烃的混合物。石油以其被称为“工业的血液”，而重要性非比寻常。它因此成为国际地缘政治的核心之一。

因此，影响原油价格的因素非常复杂。它取决于大国博弈的结果。如果各大国对当前原油价格满意，那么，他们会维持油价相对稳定，则油价会呈现反转特征。如果有大国试图打破平衡，通过油价的波动去实现自己的战略目标，则油价又会呈现出趋势特征。

另外，和黄金一样，它受全球重大事件的影响，因此，小尺度策略以跨市场套利为主。但它的跨市场套利的难度又比黄金更大。因为不同市场的原油是不同的，而地缘政治带来的影响又对不同市场显著不同。一个事件对美国的原油可能是利空，而对中国的原油可能是利好。这无疑更增加了其难度。

### 9.2.6 大豆期货

大豆期货的标的有两种，在我国被称为豆一、豆二。简单说，豆一是转基因大豆，豆二是非转基因大豆。除了我国以外，全世界生产的大豆几乎都是转基因大豆，即豆一。换言之，豆一几乎都是进口的，豆二几乎都是国产的。

豆一的主要用途是榨油，榨出的豆油供食用，榨油剩下的豆粕用于制作饲料。豆二的主要用途是食用。豆二可以对豆一进行替代，豆一不能替代豆二。全世界豆一的产量约3亿吨，中国进口约1亿吨。全世界豆二产里不足0.2亿吨，几乎全是中国产的，供食用。

中美贸易战中，大豆扮演了重要角色。美国、巴西、阿根廷是豆一的主产国，中国是最大的消费国。美国的大豆是秋季收获，巴西、阿根廷的大豆是春季收获，这就保证了我国大豆供应的连续性。但这种季节性的差异导致不同季节的合约，对应了完全不同的大豆。因此，不同到期月的合约，不宜连续化研究，这是农产品期货的一个重要特点。好在大豆易储存，问题还不大。但豆粕不易储存，这个影响就比较大，而且它会影响豆一的价格。

例如，豆粕库存高，它不易储存就意味着必须抛售，价格就会降。这必然导致豆一、豆油价格的下降，而又以豆油价格下降最甚。而豆一又无法替代豆二，所以豆二并不会因此降价。

在中美贸易顺利的时候，大豆价格主要依赖于美洲的天气，美豆与中国的大豆同向变化；而当中美发生贸易战时，则美豆与中国的大豆反向变化。

一般而言，农产品的价格也呈尖峰厚尾的特性——大部分时间是反转特性比较强，发生极端事件时出现趋势。但这种特性是否明显，又受政策影响。比如大豆对我国来说，不是战略物资，但又特别重要，所以国家会允许它在一定范围内浮动。因为豆粕主要是用于生产猪饲料，而猪价对CPI影响很大。所以，大豆价格要为稳定猪肉价格服务——只要它不影响猪肉价格，它就可以自由浮动；但凡它影响了猪肉价格，国家就会调控。那么，一旦国家调控，它一般会走向反转；如果国家不调控，则它倾向于趋势。因此，研究大豆期货时，也需要考虑政策影响。

其他农产品，如小麦、玉米、粳稻这类易储存的主粮，则是以反转为主要特征；苹果这类不太重要，又难储存的经济作物，只要不影响CPI，则以趋势为主要特征。

### 9.2.7 镍期货

镍期货的标的是镍，它是有色金属的代表。有色金属在工业上用途非常广。它的价格波动受政策影响较小，也不易被调控。它主要受经济周期的影响，因此，它以趋势为主要特征。

另外，有色金属会受科技进步的影响。以镍为例，镍主要用于生产不锈钢，但它又可以用于生产电池。在新能源革命的进程中，它的用量很可能会大幅提升。但市场完全无法预计镍用量会提升多少。也许，因为技术改进，未来电池可能就不再使用镍了，这也是有可能的。这也就导致其波动较大。而这种波动的可预测性又比较强，因此，在各个尺度上，趋势性都比较明显。

### 9.2.8 塑料期货

塑料期货的标的有很多种，比如聚乙烯、聚丙烯、聚氯乙烯等。塑料是石油化工的产物，所以受原油价格的影响非常大。另外，塑料在工业上用途也非常广，因此也受经济周期影响很大。再者，石油化工产业链较长，任何一个环节出现问题，都可能导致产业链局部价格的暴涨暴跌。因此，总体说来，塑料期货价格以趋势性为主。

近年来，电商发展迅速，快递包裹量飞速上升，对塑料包装的需求也大幅提升。一方面，这也意味着社会进步会给塑料带来趋势性行情；另一方面，它对CPI的影响也逐步提升，一旦塑料价格上涨影响了CPI，国家也肯定会出手调控。这意味着未来它可能也会逐步具有反转特性。

## 9.3 典型策略

### 9.3.1 一主多辅趋势交易

这是CTA策略最早期的形式。它的特点是以一个主策略为核心，决定开平仓的时机，同时辅以多种优化手段。

主策略以突破、均线为主要工具，以判断并顺应趋势为主要目标，有的还会浮盈加仓。典型代表有网格突破、海龟交易等。

但主策略往往对趋势的判断并不精确，在回撤时处置也不及时，因此，人们会加入各种辅助优化手段。比如，跟踪止损、特定条件不开仓等。

### 9.3.2 多策略趋势交易

这是进阶的形式。过去一主多辅的模式对主策略依赖太大。就国内而言，CTA策略在2013年有长达半年的连续大幅回撤，使人们第一次开始正视业绩曲线的平滑程度了。于是人们开始同时使用多个主策略，同时也不放弃辅助优化手段。这样，多个主策略之间分散了风险。注意，这里分散的是策略失效的风险，而对市场风险并没有太显著的效果。同时，人们也使用关联品种的走势来构造主策略。由此，主策略的来源更丰富了，业绩曲线也更平稳了。

### 9.3.3 多策略交易

因为反转做起来相对较难，收益也不是特别吸引人，因此，它是最后被引入的策略。在2017年，国内绝大部分CTA基金亏损，这使人们第一次开始正视反转策略对业绩曲线的平滑作用。反转策略不是为了分散策略失效的风险而存在，而是为了分散市场风险而存在。有了反转策略之后，CTA策略整体稳健多了。
